```python
#libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/raw/patient_encounters_2023.csv")
```

```python
#shape
df.shape
```

```python
#columns
df.columns
cols = df.columns.tolist()

for col in cols:
    print(col)
```

```python
#dtypes for each column
df.dtypes
```

```python
#all null values
df.replace(["?", "NA", "N/A", "null", "None", ""], np.nan, inplace=True)

df.isnull().sum()
```

```python
#other nulls
fake_nulls = ["?", "NA", "N/A", "null", "None", ""]

for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].isin(fake_nulls).sum())
```

```python
#empty strings/spaces
for col in df.select_dtypes(include='object'):
    print(f"\nColumn: {col}")
    print((df[col].str.strip() == "").sum())
```

```python
#unique values
for col in df.columns:
    print(f"\n{col}")
    print(df[col].unique()[:10]) 
```

```python
#Value Counts (Best for catching hidden issues)
for col in df.columns:
    print(f"\n===== {col} =====")
    print(df[col].value_counts(dropna=False))
```

```python
#Detect Strings That Should Be Numeric
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            pd.to_numeric(df[col])
        except:
            print(f"{col} may contain non-numeric or dirty values")
```

```python
#target variable distribution/class imbalance
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(normalize=True) * 100)
```
